```python
import graphviz
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('20220328-or-eng-full.csv')
# removed_cols = ['Postoperative Olanzapine',
#                 'Postoperative Fluphenazine', 'Postoperative Flupentixol']
# data = data.drop(removed_cols, axis=1)

# 将特征和目标分开
features = data.drop('Label', axis=1).values
target = data['Label'].values

train_features, test_features, train_target, test_target = train_test_split(
    features, target, test_size=0.3, random_state=42)
```

```python
# 创建决策树模型，设置最大深度为3
model = DecisionTreeClassifier(max_depth=10)

# 拟合模型
model.fit(train_features, train_target)

# 生成Graphviz格式的决策树图
dot_data = export_graphviz(model, out_file=None,
                           feature_names=data.columns[:-1],
                           class_names=['0', '1'],
                           filled=True, rounded=True,
                           special_characters=True)
```

```python
# 展示决策树图
graph = graphviz.Source(dot_data)
graph.view()

# high resolution
graph.render('decision_tree', format='png')

# Test accuracy
print(model.score(test_features, test_target))

sensitivity = model.score(test_features[test_target == 1],
                          test_target[test_target == 1])
specificity = model.score(test_features[test_target == 0],
                          test_target[test_target == 0])
print('sensitivity: ', sensitivity)
print('specificity: ', specificity)
```
